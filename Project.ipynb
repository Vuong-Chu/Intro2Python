```python
import pandas as pd

import numpy as np
```

```python
data = pd.read_csv("SucidesIndia.csv")
```

```python
data

data["State"].unique()
```

```python
data = data[(data["State"]!="Total (All India)") | (data["State"]!="Total (States)") | (data["State"]!="Total (Uts)")]

data
```

```python
data["Type_code"].unique()
#data.drop_duplicates(subset=list(data.columns)[0:6],inplace=True)
```

```python
data1 = data[data["Type_code"]=="Causes"].copy()

data1 = data1.pivot_table(index=["State","Year","Gender","Age_group"],columns=["Type"],values=["Total"],aggfunc=np.sum)

data1.columns = data1.columns.to_flat_index()

data1 = data1.reset_index()

data1 = pd.melt(data1,id_vars=["State","Year","Gender","Age_group"], var_name="Causes", value_name="Value")

data1 = data1.dropna()

data1 = data1[data1["Value"]!=0]

data1.set_index(["State","Year","Gender","Age_group"])
```
